```python

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sheets_dict = pd.read_excel('bode1/EVOCARD.xlsx', sheet_name=None)
```

```python
for sheet_name, df in sheets_dict.items():
    if sheet_name == 'CARD_FACT':
        fact_df = df.copy()
    elif sheet_name == "CARD_DIM":
        dim_df =  df.copy()
```

```python
# merge 2 sheets
merged_df = pd.merge(fact_df, dim_df, on=['CIF hash', 'CARD_NBR hash'], how="inner", suffixes=('_left','_right'))

# convert transaction time
merged_df['TXN_TM'] = pd.to_datetime(merged_df['TXN_TM'])

# sort by transaction time
merged_df = merged_df.sort_values(by='TXN_TM', ignore_index=True)

# drop unsuccessful transaction
merged_df = merged_df[~merged_df["TXN_SRC_ID"].isna()]
```

```python
def split_period_time(df, time_col):
    df['year'] = df[time_col].dt.year
    df['month'] = df[time_col].dt.month
    df['day'] = df[time_col].dt.day

    # Create a mask for rows where date falls between 22nd of one month and 21st of the next month
    mask = ((df[time_col].dt.day >= 22) & (df[time_col].dt.month == df['month'])) | \
        ((df[time_col].dt.day <= 21) & (df[time_col].dt.month == df['month'] + 1))
    interval_1 = df[mask]
    interval_2 = df[~mask]
    return interval_1, interval_2

split_period_time(merged_df, "TXN_TM")
```

```python
def filter_add_debt(df: pd.DataFrame):
    add_debt_df = df.loc[((df["TML_WEB_AP_NM"] == "OPTP0000") | (df["TML_WEB_AP_NM"] == "OPTP0119")), :]
    return add_debt_df

add_debt_df = filter_add_debt(merged_df)
add_debt_df.head(5)
```

```python
def filter_pay_debt(df: pd.DataFrame):
    pay_debt_df = df.loc[((df["TML_WEB_AP_NM"] == "OPTP0028") | (df["TML_WEB_AP_NM"] == "OPTP1501")), :]
    return pay_debt_df

pay_debt_df = filter_pay_debt(merged_df)
pay_debt_df.head()
```

```python
def calculate_debt_per_period(df: pd.DataFrame):
    mask = ((df['TXN_TM'].dt.day <= 21) & (df['TXN_TM'].dt.month == df['month'])) | \
       ((df['TXN_TM'].dt.day >= 22) & (df['TXN_TM'].dt.month == df['month'] - 1))
    grouped_df = df[mask].groupby(['CIF hash', 'CARD_NBR hash', 'CREDIT_LIMIT', 'year', 'month']).agg({
        # Add aggregation functions for other columns as needed
        'DSC': 'count',  # Example aggregation: count the number of transactions
        'NET_CASH_FLOW_AMT_LCY': 'sum'  # Example aggregation: sum of cash flow amounts
    })
    return grouped_df

report_debt = calculate_debt_per_period(add_debt_df)
report_debt.head(10)
```
